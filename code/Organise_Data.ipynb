```python
import pandas as pd
import numpy as np
```

```python
anime = pd.read_csv("../DataSet/anime.csv") 
rating = pd.read_csv("../DataSet/rating.csv") 
```

```python
anime.head()
```

```python
rating.head()
```

```python
anime.isnull().sum()
```

```python
anime['genre'].fillna(method='ffill', inplace=True)

anime['type'].fillna(method='ffill', inplace=True)

anime['rating'].fillna(round(anime['rating'].mean(), 2), inplace=True)
```

```python
rating.isnull().sum()
```

```python
anime.to_csv('../DataSet/anime.csv', index=False)
```

```python
mean_ratings = rating.groupby('anime_id')['rating'].mean().reset_index()
```

```python
mean_ratings.head()
```

```python
#users row in ratings csv 
count_users = rating.groupby('anime_id').size().reset_index(name='count')
```

```python
count_users.head()
```

```python
# Rename the 'rating' column to 'user_rating'
mean_ratings.rename(columns={'rating': 'user_rating'}, inplace=True)
```

```python
# Merge the anime dataframe with mean ratings based on anime_id
merged_df = pd.merge(anime, mean_ratings, on='anime_id', how='left')
merged_df=pd.merge(merged_df,count_users, on='anime_id',how='left')

# Save the merged dataframe back to a new CSV file
merged_df.to_csv('../DataSet/updated_anime.csv', index=False)
```

```python
updated_anime=pd.read_csv("../DataSet/updated_anime.csv") 
```
