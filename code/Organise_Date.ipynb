```python
import pandas as pd
import numpy as np
```

```python
anime = pd.read_csv("anime.csv") 
rating = pd.read_csv("rating.csv") 
```

```python
anime.head()
```

```python
rating.head()
```

```python
anime.isnull().sum()
```

```python
anime['genre'].fillna(method='ffill', inplace=True)

anime['type'].fillna(method='ffill', inplace=True)

anime['rating'].fillna(round(anime['rating'].mean(), 2), inplace=True)
```

```python
rating.isnull().sum()
```

```python
anime.to_csv('anime.csv', index=False)
```

```python
mean_ratings = rating.groupby('anime_id')['rating'].mean().reset_index()
```

```python
mean_ratings.head()
```

```python
#users row in ratings csv 
count_users = rating.groupby('anime_id').size().reset_index(name='count')
```

```python
count_users.head()
```

```python
# Rename the 'rating' column to 'user_rating'
mean_ratings.rename(columns={'rating': 'user_rating'}, inplace=True)
```

```python
# Merge the anime dataframe with mean ratings based on anime_id
merged_df = pd.merge(anime, mean_ratings, on='anime_id', how='left')
merged_df=pd.merge(merged_df,count_users, on='anime_id',how='left')

# Save the merged dataframe back to a new CSV file
merged_df.to_csv('updated_anime.csv', index=False)
```

```python
updated_anime=pd.read_csv("updated_anime.csv") 
```

```python
updated_anime.head()
```

```python
#approx. the user rating column
updated_anime['user_rating']=updated_anime['user_rating'].round(2)
updated_anime['count']=updated_anime['count'].fillna(0).astype(int)
# Save the merged dataframe back to a new CSV file
updated_anime.to_csv('updated_anime.csv', index=False)
updated_anime.head()
```

```python
#Top anime by communities 
top20_anime= updated_anime[updated_anime['members']>100000]
top20_anime= top20_anime[top20_anime['count']>10000]

# Sort the DataFrame by 'members' and 'user_rating' in descending order to get the top 20
top20_anime = top20_anime.sort_values(by=['user_rating'], ascending=False).head(20)

# Create a table for the top 20 anime with 'anime_name', 'user_rating', and 'members' columns
top20_anime = top20_anime[['name', 'user_rating', 'members','count']]

top20_anime.head()
```
